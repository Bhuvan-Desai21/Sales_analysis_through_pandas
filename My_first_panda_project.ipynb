```python
import pandas as pd
from itertools import combinations
from collections import Counter
import matplotlib.pyplot as plt


all_data = all_data.dropna(subset=["Order ID"])


bought_together = all_data[all_data.duplicated(subset=["Order ID"], keep=False)]


product_combinations = []
for order_id, group in bought_together.groupby("Order ID"):
    products = group["Product"].unique()
    product_combinations.extend(combinations(products, 2)) 


combination_counts = Counter(product_combinations)


most_common_pairs = combination_counts.most_common(10) 


pairs = [f"{pair[0]} & {pair[1]}" for pair, count in most_common_pairs]
counts = [count for pair, count in most_common_pairs]


plt.figure(figsize=(12, 8))
plt.barh(pairs, counts, color='skyblue')
plt.xlabel('Number of Times Bought Together', fontsize=14)
plt.ylabel('Product Pairs', fontsize=14)
plt.title('Top 10 Most Commonly Bought Together Product Pairs', fontsize=16)
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("bought_together.jpg")
plt.show()
```
